### TF-IDF   

```python
import sqlite3
import pandas as pd
```

```python
def get_data_from_database():
    conn = sqlite3.connect('../data/db_steam.db')
    df = pd.read_sql('SELECT * FROM reviews', conn)
    conn.close()
    return df

df = get_data_from_database()
```

```python
#tfidf of reviews
from sklearn.feature_extraction.text import TfidfVectorizer

# Create an instance of TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit the vectorizer to the text data and transform it into TF-IDF vectors
tfidf_vectors = tfidf_vectorizer.fit_transform(df['review_clean'])

# Convert the TF-IDF vectors to an array
tfidf_array = tfidf_vectors.toarray()

# Get the feature names (words)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a DataFrame for better visualization (optional)
import pandas as pd
tfidf_df = pd.DataFrame(tfidf_array, columns=feature_names)

tfidf_array.shape
```

```python
#tfidf of reviews
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import csr_matrix

# Create an instance of TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

tfidf_vectors = tfidf_vectorizer.fit_transform(df['review_clean'])

feature_names = tfidf_vectorizer.get_feature_names_out()

tfidf_csr_matrix = csr_matrix(tfidf_vectors)

tfidf_dataframe = pd.DataFrame.sparse.from_spmatrix(tfidf_csr_matrix, columns=feature_names)

tfidf_dataframe
```

```python
sorted_tfidf_dataframe = tfidf_dataframe.sort_values(by=feature_names[0], ascending=False)

top_document = sorted_tfidf_dataframe.iloc[0]

top_words = top_document.sort_values(ascending=False).head(20)
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['review_clean'])
feature_names = tfidf_vectorizer.get_feature_names_out()

tfidf_scores_df = pd.DataFrame({'Wort': feature_names, 'TF-IDF-Score': tfidf_matrix.mean(axis=0).tolist()[0]})
tfidf_scores_df = tfidf_scores_df.sort_values(by='TF-IDF-Score', ascending=False)
tfidf_scores_df.head(20)
```
